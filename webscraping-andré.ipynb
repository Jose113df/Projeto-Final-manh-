```python
import pandas as pd

url = 'https://www.terra.com.br/esportes/futebol/brasileiro-serie-a/tabela/#google_vignette'

tabela = pd.read_html(url, encoding='utf-8')

type(tabela)
```

```python
### Passo 1.19: Renomear nome de uma coluna

df.rename(columns={'D':'derrotas'},inplace=True)
df.rename(columns={'V':'vitorias'},inplace=True)
df.rename(columns={'E':'empate'},inplace=True)
```

```python
df = df.drop(columns=['Times','Times.1','Times.3'])
```

```python
#Manipulação de Strings e Intro (Ex. regular)
df['times'] = df['times'].str.replace('>>','',regex=False).str.strip()
```

# Gráficos
    -Gráficos para usuário basico, exemplo: Em barra (horizontal/vertical)
    -Gráficos para usuário avançado, exemplo: dispersão

```python
df[['times','V']]

df.plot('times', 'V')
```

```python
import matplotlib as plt
```

```python
grafico_percentual_vitorias = df.groupby('times')[['percentual']].sum().sort_values('percentual')
```

```python
grafico_percentual_vitorias.plot(kind='barh',figsize=(14,10),color='red')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que 'df' seja seu DataFrame
# Agrupando por 'times' e somando as colunas 'D', 'V', 'E'
grafico_vitorias_derotas_empate = df.groupby('times')[['D', 'V', 'E']].sum()

# Exibindo as somas de vitórias, derrotas e empates
print(grafico_vitorias_derotas_empate)

# Para o gráfico de pizza, você deve escolher uma coluna específica
# Por exemplo, vamos fazer o gráfico de pizza para vitórias 'V'
grafico_vitorias_derotas_empate['V'].plot(kind='pie', figsize=(14,10), autopct='%1.1f%%')

plt.title('Distribuição das Vitórias por Time')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponha que 'df' seja o seu DataFrame e que você queira filtrar um time específico.
time_especifico = input("Difite o nome do time que deseja ver o gráfico: ")

# Filtrando o DataFrame para o time específico
dados_time = df[df['times'] == time_especifico]

# Somando as vitórias, derrotas e empates para o time específico
soma_dados_time = dados_time[['derrotas', 'vitorias', 'empate']].sum()

# Criando o gráfico de pizza
soma_dados_time.plot(kind='pie', figsize=(14, 10), autopct='%1.1f%%', colors=['red', 'green', 'grey'])

plt.title(f'Distribuição de Resultados do Time {time_especifico}')
plt.ylabel('')  # Remove o rótulo do eixo y
plt.show()

```
